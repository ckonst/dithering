```python
from PIL import Image
from scipy import signal
import numpy as np
```

```python
white_noise = np.round(np.random.rand(300, 300) * 255.0).astype(np.uint8)
white_noise_image = Image.fromarray(white_noise)
display(white_noise_image)
```

```python
def butter_highpass_filter(input_sig, cutoff, fs, order=1):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = signal.butter(order, normal_cutoff, btype='high', analog=False)
    return signal.filtfilt(b, a, input_sig)
```

```python
blue_noise = butter_highpass_filter(white_noise, cutoff=100, fs=300)
blue_noise_image = Image.fromarray(blue_noise).convert('RGB')
display(blue_noise_image)
```

```python
def butter_bandpass_filter(input_sig, low_cut, high_cut, fs, order=1):
    nyq = 0.5 * fs
    low, high = low_cut / nyq, high_cut / nyq
    b, a = signal.butter(order, (low, high), btype='bandpass')
    return signal.filtfilt(b, a, input_sig)
```

```python
green_noise = butter_bandpass_filter(white_noise, low_cut=5, high_cut=75, fs=300)
green_noise_image = Image.fromarray(green_noise).convert('RGB')
display(green_noise_image)
```
